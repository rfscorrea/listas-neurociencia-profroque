```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
def solucoes(v0, t):
    w0 = -5/8

    def dALLdt(X,t):
        v, w = X

        dvdt = v - (v**3)/3 - w
        dwdt = 0.08 * (v + 0.7 - 0.8 * w)

        return dvdt, dwdt


    X = odeint(dALLdt, [v0, w0], t)
    v = X[:,0]
    w = X[:,1]

    return v, w
```

```python
t = np.arange(0,40,0.1)

v1, w1 = solucoes(-0.8, t)
v2, w2 = solucoes(-0.65, t)
v3, w3 = solucoes(-0.64, t)
v4, w4 = solucoes(0, t)
```

```python
plt.style.use('bmh')

plt.subplot(2,1,1)
plt.plot(t, v1, label='v0 = -0.8')
plt.plot(t, v2, label='v0 = -0.65')
plt.plot(t, v3, label='v0 = -0.64')
plt.plot(t, v4, label='v0 = 0')

plt.ylabel('v')
plt.legend()

plt.subplot(2,1,2)
plt.plot(t, w1, label='v0 = -0.8')
plt.plot(t, w2, label='v0 = -0.65')
plt.plot(t, w3, label='v0 = -0.64')
plt.plot(t, w4, label='v0 = 0')

plt.ylabel('w')
plt.xlabel('t')
plt.legend()

plt.show()
```
