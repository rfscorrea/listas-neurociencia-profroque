```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
def solucoes(I, t):
    w0 = -5/8
    v0 = -6/6

    def dALLdt(X,t):
        v, w = X

        dvdt = v - (v**3)/3 - w + I
        dwdt = 0.08 * (v + 0.7 - 0.8 * w)

        return dvdt, dwdt


    X = odeint(dALLdt, [v0, w0], t)
    v = X[:,0]
    w = X[:,1]

    return v, w


def nulclina_v(v, I):
    return v - (v**3)/3 + I


def nulclina_w(v):
    a = 0.7
    b = 0.8

    return (v+a) / b
```

```python
t = np.arange(0,40,0.1)

v1, w1 = solucoes(1, t)
v2, w2 = solucoes(1.5, t)

v = np.arange(-2,2,0.01)
w_I1 = nulclina_v(v, 1)
w2_nulclina = nulclina_w(v)
w_I2 = nulclina_v(v, 1.5)
```

```python
plt.style.use('bmh')
```

```python
plt.figure()

plt.subplot(2,2,1)
plt.title('I = 1')
plt.plot(v, w_I1, label='dv/dt = 0')
plt.plot(v, w2_nulclina, label='dw/dt = 0')
plt.legend()

plt.xlabel('v')
plt.ylabel('w')



plt.subplot(2,2,2)
plt.title('I = 1.5')
plt.plot(v, w_I2, label='dv/dt = 0')
plt.plot(v, w2_nulclina, label='dw/dt = 0')
plt.legend()

plt.xlabel('v')
plt.ylabel('w')


plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(t, v1, label='I = 1')
plt.plot(t, v2, label='I = 1.5')

plt.ylabel('v')
plt.legend()

plt.subplot(2,1,2)
plt.plot(t, w1, label='I = 1')
plt.plot(t, w2, label='I = 1.5')

plt.ylabel('w')
plt.xlabel('t')
plt.legend()

plt.show()
```
