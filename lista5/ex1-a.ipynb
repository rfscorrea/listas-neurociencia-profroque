```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def nulclina_v(v):
    return v - (v**3)/3

def nulclina_w(v):
    a = 0.7
    b = 0.8

    return (v+a) / b
```

```python
v = np.arange(-2, 2, 0.01)
w1 = nulclina_v(v)
w2 = nulclina_w(v)
```

```python
plt.style.use('bmh')

plt.plot(v, w1, label='dv/dt = 0')
plt.plot(v, w2, label='dw/dt = 0')

plt.xlabel('v')
plt.ylabel('w')

plt.legend()
plt.show()
```

```python
for i in range(len(v)):
    if w1[i] == w2[i]:
        print(i)
```
